```python
import re
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))  #unique stopwords
lemmatizer = WordNetLemmatizer()
```

```python
text = 'I love this movie !!!'
print (f'Original Text: {text}')
```

```python
def preprocess_text(text):
    text = text.lower()  #lowercasing

    text = text.translate(str.maketrans('','',string.punctuation))  #remove puncation

    text = re.sub (r'[^\w\s]','',text)  #remove special characters/emojis
    #\w means remove extra spaces

    tokens = text.split()  #tokenization

    tokens = [word for word in tokens if word not in stop_words]  #stopwords removal

    tokens = [lemmatizer.lemmatize(word) for word in tokens]  #lemmatization

    return tokens
```

```python
processes = preprocess_text(text)
print (F"Processed Text : {processes}")
```
