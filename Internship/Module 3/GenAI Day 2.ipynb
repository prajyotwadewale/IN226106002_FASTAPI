```python
import re
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english')) # unique stopwords
lemmatizer = WordNetLemmatizer()
```

```python
text = 'I love this movie!!!!'
print(f'Original Text: {text}')
```

```python
def preprocess_text(text):
  #1. Lowercasing
  text = text.lower()

  #2. Remove punctuations
  text = text.translate(str.maketrans('', '', string.punctuation))

  #3.Remove Special Characters/Emojis
  text = re.sub(r'[^\w\s]', '', text)

  #4. Tokenization
  tokens = text.split()

  #5. Stopwords removal
  tokens = [word for word in tokens if word not in stop_words]

  #6. Lemmatization
  tokens = [lemmatizer.lemmatize(word) for word in tokens]

  return tokens
```

```python
processed = preprocess_text(text)
print(f'Processed Text: {processed}')
```

```python
text = [ 'I hate this movie!!',
        'This course is very helpful',
         'Worst experience ever!!!',
         ' I am learning NLP and it is amazaing ']
```

```python
for i in text:
  print(f'Original: {i}')
```

```python
for i in text:
  print(f'Processed: {preprocess_text(i)}')
```

```python
def tokens_to_sentence(tokens):
  return ' '.join(tokens)
```

```python
example = preprocess_text('This course is Amazing!!!!')
print(example)
```

```python
tokens_to_sentence(example)
```
